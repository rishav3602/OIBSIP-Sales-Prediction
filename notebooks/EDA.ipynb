```python
import pandas as pd
```

```python
## Data Ingestions step
data=pd.read_csv('data/aad.csv')
data.head()
```

```python
## making a copy of data so that it do not harm my original dataset.
df = data.copy()
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
## we will check for null values in our dataset.
df.isnull().sum()
```

```python
## we do not have any null values in our dataset and we are good to go...!!!
```

```python
## Lets drop the  unnamed column
df=df.drop(labels=['Unnamed: 0'],axis=1)
df.head()
```

```python
## check for duplicated records
df.duplicated().sum()
```

```python
## we do not have any duplicate values in our dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
x=0
for i in df.columns:
    sns.histplot(data=df,x=i,kde=True)
    print('\n')
    plt.show()
```

```python
## correlation
sns.heatmap(df.corr(),annot=True)
```
